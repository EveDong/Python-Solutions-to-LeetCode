# 7. Reverse Integer

```python
# range 2^31-1=2147483647,-2^31=-2147483648
class Solution:
    def reverse(self, x: int) -> int:
        result = int(str(abs(x))[::-1])
        if x < 0 and -result > -(2 ** 31):
            return -result
        elif x >= 0 and result < 2 ** 31 - 1:
            return result
        else:
            return 0
```

```python
class Solution:
    def reverse(self, x: int) -> int:
        if abs(x) < 10:
            return x
        if x < 0:
            temp = -int(str(x)[:0:-1])
        else:
            temp = int(str(x)[::-1])
        return temp if -(2 ** 31) < temp < 2 ** 31 - 1 else 0
```

```python
# python3中， /是精确除法，//是向下取整除法，%是求模，四舍五入取整round, 向零取整int, 向下和向上取整函数math.floor, math.ceil
class Solution:
    def reverse(self, x: int) -> int:
        y, res = abs(x), 0
        # 则其数值范围为 [−2^31,  2^31 − 1]
        boundry = (1 << 31) - 1 if x > 0 else 1 << 31
        while y != 0:
            res = res * 10 + y % 10
            if res > boundry:
                return 0
            y //= 10
        return res if x > 0 else -res
```

# 8. String to Integer (atoi)

```python
class Solution:
    def myAtoi(self, s: str) -> int:
        # without digit
        if not "".join(filter(str.isnumeric, s)):
            return 0
        # first character
        i = 0
        if not ("0" <= s[i] <= "9" or s[i] == "-" or s[i] == "+" or s[i] == " "):
            return 0
        # ' '
        if s[i] == " ":
            while i < len(s):
                if s[i] != " ":
                    break
                i += 1
        # sign
        if s[i] == "-" or s[i] == "+":
            if not "0" <= s[i + 1] <= "9":
                return 0
            result = s[i]
            s = s[i + 1 :]
        else:
            result = ""
            s = s[i:]
        # digit
        for item in s:
            if "0" <= item <= "9":
                result += item
            else:
                break
        # sign + digit
        if result == "":
            return 0
        # interval
        result = int(result)
        if result < -(2 ** 31):
            return -(2 ** 31)
        elif result > 2 ** 31 - 1:
            return 2 ** 31 - 1
        else:
            return result
```

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if needle == "":
            return 0
        if len(haystack) < len(needle):
            return -1
        i = 0
        j = 0
        while i < len(haystack) and j < len(needle):
            if haystack[i] != needle[j]:
                i = i - j + 1
                j = 0
            else:
                i += 1
                j += 1
        if j == len(needle):
            return i - j
        return -1
```

# 38. Count and Say

```python
class Solution:
    def countAndSay(self, n: int) -> str:
        s = "1"
        if n == 1:
            return s
        for i in range(2, n + 1):
            temp = []
            j = 0
            for k in range(1, len(s)):
                if s[j] != s[k]:
                    temp.append(s[j:k])
                    j = k
            temp.append(s[j:])
            s = ""
            for i in temp:
                s += str(len(i))
                s += i[0]
        return s
```

```python
class Solution:
    def countAndSay(self, n: int) -> str:
        arr = ["1"]
        for i in range(n - 1):
            temp = ""
            j = 0
            while j < len(arr[i]):
                num = arr[i][j]
                time = 0
                while j < len(arr[i]) and num == arr[i][j]:
                    j += 1
                    time += 1
                temp = temp + str(time) + num
            arr.append(temp)
        return arr[n - 1]
```

# 125. Valid Palindrome

```python
# pythonic
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = "".join(filter(str.isalnum, s)).lower()
        return s == s[::-1]
```

```python
# two pointers
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        i, j = 0, len(s) - 1
        while i <= j:
            if not s[i].isalnum():
                i += 1
                continue
            if not s[j].isalnum():
                j -= 1
                continue
            if s[i] != s[j]:
                return False
            i += 1
            j -= 1
        return True
```

```python
# a" <= i <= "z" or "0" <= i <= "9"
class Solution:
    def isPalindrome(self, s: str) -> bool:
        if len(s) < 2:
            return True
        else:
            temp = []
            for i in s.lower():
                if "a" <= i <= "z" or "0" <= i <= "9":
                    temp.append(i)
            return temp[::-1] == temp
```

# 242. Valid Anagram

```python
# lowercase alphabets
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        d = {i: 0 for i in "abcdefghijklmnopqrstuvwxyz"}
        for i in s:
            d[i] += 1
        for i in t:
            d[i] -= 1
        for i in d:
            if d[i] != 0:
                return False
        return True
```

```python
# hash
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        if len(s) == 0:
            return True
        d = {}
        for i in s:
            if i not in d:
                d[i] = 1
            else:
                d[i] += 1
        for i in t:
            if i not in d:
                return False
            else:
                d[i] -= 1
        for i in d.values():
            if i != 0:
                return False
        return True
```

# 387. First Unique Character in a String

```python
# abcdefghijklmnopqrstuvwxyz
class Solution(object):
    def firstUniqChar(self, s: str) -> int:
        # 先假设最小索引为最后的字符索引
        min_unique_char_index = len(s)
        # 已知字符串由小写字母构成，则遍历a-z
        for c in "abcdefghijklmnopqrstuvwxyz":
            i = s.find(c)
            # 分别从目标的字符串头和字符串尾查找对应字母的索引；如果两索引相等，则说明是单一字符
            if i != -1 and i == s.rfind(c):
                # 更新最新的最小索引
                min_unique_char_index = min(min_unique_char_index, i)
        # 如果返回值不为最后字符的索引，则返回最小索引值
        # 否则，根据题意，返回-1
        return min_unique_char_index if min_unique_char_index != len(s) else -1
```

```python
# enumerate
class Solution(object):
    def firstUniqChar(self, s: str) -> int:
        # 计算字符出现次数
        dic = {c: s.count(c) for c in set(s)}
        # 找到并返回首个满足出现次数为一的字符
        for i, c in enumerate(s):
            if dic[c] == 1:
                return i
        return -1
```

```python
# hash OrderedDict
class Solution:
    def firstUniqChar(self, s: str) -> int:
        if len(s) == 0:
            return -1
        if len(s) == len(set(s)):  # len(s)=1 or no repeating character
            return 0
        d = collections.OrderedDict()
        for i in range(len(s)):
            if s[i] not in d:
                d[s[i]] = 0
            else:
                d[s[i]] += 1
        for key, value in d.items():
            if value == 0:
                return s.index(key)
        return -1
```
